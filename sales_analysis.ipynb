```python
# Step 1 - Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2 - Load CSV file
# Make sure "sales.csv" is in your working folder or provide full file path
df = pd.read_csv("sales.csv")

# Step 3 - View first few rows
print("First 5 rows of the dataset:")
print(df.head())

# Step 4 - Check column names and data info
print("\nDataset Info:")
print(df.info())

# Step 5 - Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Step 6 - Total sales by region
sales_by_region = df.groupby("Region")["Sales"].sum()
print("\nTotal Sales by Region:")
print(sales_by_region)

# Step 7 - Total quantity sold by product
quantity_by_product = df.groupby("Product")["Quantity"].sum()
print("\nTotal Quantity by Product:")
print(quantity_by_product)

# Step 8 - Plot sales by region
sales_by_region.plot(kind="bar", title="Total Sales by Region", color="skyblue")
plt.ylabel("Total Sales")
plt.show()

# Step 9 - Plot quantity sold by product
quantity_by_product.plot(kind="pie", autopct="%1.1f%%", title="Quantity Sold by Product")
plt.ylabel("")  # Hides the label
plt.show()

# Step 10 - Monthly sales trend
# Convert Date column to datetime type
df["Date"] = pd.to_datetime(df["Date"])

# Group by month and sum Sales
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["Sales"].sum()

# Plot trend
monthly_sales.plot(kind="line", marker="o", title="Monthly Sales Trend", color="green")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
